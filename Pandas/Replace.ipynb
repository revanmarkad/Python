# REPLACE

```python
import pandas as pd

var=pd.read_csv("Country.csv",encoding='ISO-8859-1')
print(var)
```

```python
var = var.replace(to_replace='a', value='b')
print(var)
```

```python
var11 = var.replace([1,2,3,4,5,6,7],22)
print(var11)
```

```python
var3 = var.replace("[A-Za-z,]",'python',regex=True)
print(var3)
```
