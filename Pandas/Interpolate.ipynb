# convert xlsx to csv

```python
import pandas as pd

# Read the Excel file
xlsx_file = pd.read_excel("Amaravati-2025.xlsx")

# Save the DataFrame to a CSV file
xlsx_file.to_csv("Amravati.csv", index=False)
```
